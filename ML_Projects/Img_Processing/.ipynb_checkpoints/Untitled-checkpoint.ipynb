```python
"""
For this project, I have used pyplot from matplotlib to display the image being
processed. I have also used the skimage library which is the workhorse for all of
the image processing. The goal of this algorithm is to count the number of stars
in a night sky image.
"""

from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray
from skimage.io import imread
from math import sqrt
import glob
```

```python
sample_img = glob.glob("night_sky.jpg")
im = imread(sample_img, as_gray=True)
plt.imshow(im)
plt.show()
```

```python
blobs_log = blob_log(im, max_sigma=30, num_sigma=10, threshold=.1)
blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
numrows = len(blobs_log)
print("Number of stars counted: ", numrows)
```

```python
fig, ax = plt.subplots(1, 1)
plt.imshow(im)
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r+5, color='lime', linewidth=2, fill=False)
    ax.add_patch(c)
```
